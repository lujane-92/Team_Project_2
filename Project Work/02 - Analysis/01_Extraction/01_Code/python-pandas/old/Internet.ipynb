# 2.9. Number of internet users by country

## 2.9.1. Country Data

```python
url = 'https://ourworldindata.org/grapher/data/variables/99383.json?v=3'
Get_Data = requests.get(url).json()
#pprint(get_data)
```

```python
Country_ID = []
Country_Name = []
Country_Code = []

for E in Get_Data['entityKey']:
    Country_ID.append(E)
    Country_Name.append(Get_Data['entityKey'][E]['name'])
    Country_Code.append(Get_Data['entityKey'][E]['code'])
```

```python
#print(Country_ID)
len(Country_ID)
```

```python
#print(Country_Code)
len(Country_Code)
```

```python
#print(Country_Name)
len(Country_Name)
```

```python
Country_df = pd.DataFrame({
    'Country_ID'    : Country_ID,
    'Country_Code'  : Country_Code,
    'Country_Name'  : Country_Name
})
```

```python
Country_df
```

```python
# Export to CSV
Country_df.to_csv("../../02_Data_Output/Internet_Users_Country_Countries_df.csv",
                   index=False, 
                   header=True
                  )
```

## 2.9.2 Values

```python
Country_ID = []
Number_Of_Internet_Users_By_Country = []
Country_Years = []


IDs      = Get_Data['variables']['99383']['entities'] 
Values   = Get_Data['variables']['99383']['values']
Years    = Get_Data['variables']['99383']['years']
    
    
for I in IDs:
    Country_ID.append(I)

for V in Values:
    Number_Of_Internet_Users_By_Country.append(V)
    
for Y in Years:
    Country_Years.append(Y)
```

```python
#print(Country_ID)
```

```python
Country_Values_df = pd.DataFrame({
    'Country_ID'     : Country_ID,
    'Country_Years'  : Country_Years,
    'Number_Of_Internet_Users_By_Country'  : Number_Of_Internet_Users_By_Country
})
```

```python
Country_Values_df
```

```python
# Export to CSV
Country_Values_df.to_csv("../../02_Data_Output/Internet_Users_Country_Values_df.csv",
                   index=False, 
                   header=True
                  )
```

# 4. URL: http://api.worldbank.org/v2/

```python
# Dependencies
import pandas as pd
from pandas import DataFrame
import requests
import json
from pandas.io.json import json_normalize

url = "http://api.worldbank.org/v2/"
format = "json"

# Get country information in JSON format
countries_response = requests.get(f"{url}country/?per_page=304&format={format}").json()
#http://api.worldbank.org/countries/?per_page=300&format=json

# First element is general information, second is countries themselves
countries = countries_response[1]
```

```python
print (countries[0])
```

```python
# Report the names
# Name, Region, Income Level, Capital City, Longitude, Latitude
for country in countries:
    print(country["name"]) 
    print(country["region"]["value"])
```

```python
Lat_Long_Coordinates_df = DataFrame(countries)
Lat_Long_Coordinates_df
```

```python
# Export to CSV
Lat_Long_Coordinates_df.to_csv("../../02_Data_Output/Lat_Long_Coordinates.csv",
                   index=False, 
                   header=True
                  )
```
