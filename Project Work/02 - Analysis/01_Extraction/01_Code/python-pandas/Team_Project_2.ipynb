```python
# Dependencies
import pandas as pd
from pandas import DataFrame
import requests
import json
from pandas.io.json import json_normalize

url = "http://api.worldbank.org/v2/"
format = "json"

# Get country information in JSON format
countries_response = requests.get(f"{url}country/?per_page=304&format={format}").json()
#http://api.worldbank.org/countries/?per_page=300&format=json

# First element is general information, second is countries themselves
countries = countries_response[1]
```

```python
print (countries[0])
```

```python
# Report the names
# Name, Region, Income Level, Capital City, Longitude, Latitude
for country in countries:
    print(country["name"]) 
    print(country["region"]["value"])
```
